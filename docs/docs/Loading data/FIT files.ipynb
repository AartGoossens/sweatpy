*Sweatpy* has built-in support for loading FIT files.

The `read_fit()` function accepts strings, pathlib objects and other file-like objects and it returns a [pandas.DataFrame](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html).

```python
import sweat

example_fit = sweat.example_data(path='4078723797.fit')

data = sweat.read_fit(example_fit.path)

data
```
