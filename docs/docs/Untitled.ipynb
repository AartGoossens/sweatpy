```python
import pandas as pd

import sweat
```

```python
for example in sweat.examples():
    data = sweat.read_file(example.path)
    diffs = data.index.to_series().diff()
    diffs = diffs.dropna()
    if not (diffs == pd.Timedelta(seconds=1)).all():
        print(f"{example.path} IS NOT sampled at 1Hz")
    else:
        print(f"{example.path} IS sampled at 1Hz")
```
